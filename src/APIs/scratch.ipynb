```python
import plotly.graph_objects as go

# Create the data for the chart
categories = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence', 'tempo']
values = df[categories].values
mean_values = df[categories].mean().values
display(values)

# Create the plotly figure
fig = go.Figure()

for i in range(0, len(values)):
    # Add the trace
    fig.add_trace(go.Scatterpolar(
        r=values[i],
        theta=categories,
        # fill='toself',
        name=df['name'][i]
    ))
# # Add the radar trace
fig.add_trace(go.Scatterpolar(
      r=mean_values,
      theta=categories,
      fill='toself',
      mode='lines',
      line = dict(color = 'black'),
))
# Update the layout
fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
  showlegend=False
)
# fig.layout.polar.radialaxis['range'] = [0, -60]


# Show the figure
fig.show()
```

```python
import plotly.express as px

# Create the parallel coordinates plot using plotly.express
categories = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'valence', 'tempo']
df_parallel = df[categories]
fig = px.parallel_coordinates(df_parallel, color='tempo')
# add a trace to the grpah
mean_values = df[categories].mean()
fig.add_trace(
    go.Scatter(
        x=mean_values.index,
        y=mean_values.values,
        mode='lines+markers',
        name='Mean Values',
        line=dict(color='red', width=2),
        marker=dict(color='red', size=10, symbol='circle')
    )
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objs as go
import pandas as pd

# Load data
# df = pd.read_csv('my_data.csv')

mean_vals = df.mean()

df_parallel = df[categories]
# append an entry
df_parallel['idividual'] = True
df_parallel = df_parallel.append(mean_vals, ignore_index=True)
df_parallel['idividual'][-1] = False
# map a color scale to columns values

df_parallel['color'] = df_parallel.apply(lambda x: 'black' if not x['idividual'] else x['tempo'], axis=1)
df_parallel['size'] = df_parallel.apply(lambda x: 2 if not x['idividual'] else 5, axis=1)
# Create trace for the parallel coordinate graph
data = [go.Parcoords(
    line=dict(color=df_parallel['color'], showscale=False),
    size=dict(df_parallel['size']),
    dimensions=list([
        dict(label='Acousticness', values=df_parallel['acousticness']),
        dict(label='Danceability', values=df_parallel['danceability']),
        dict(label='Energy', values=df_parallel['energy']),
        dict(label='Instrumentalness', values=df_parallel['instrumentalness']),
        dict(label='Liveness', values=df_parallel['liveness']),
        dict(label='Loudness', values=df_parallel['loudness']),
        dict(label='Speechiness', values=df_parallel['speechiness']),
        dict(label='Valence', values=df_parallel['valence'])
    ])
)]


# Create layout
layout = go.Layout(
    title='My Parallel Coordinates Graph',
    height=600,
    xaxis=dict(title='Features'),
    yaxis=dict(title='Values')
)

# Create figure
fig = go.Figure(data=data, layout=layout)

# Show figure
fig.show()

```

```python
from datetime import datetime, timedelta
startTime = datetime.strptime('2022-07-03', '%Y-%m-%d')
endTime = datetime.strptime('2022-10-31', '%Y-%m-%d')
endTime = datetime.now().date().strftime("%Y-%m-%d")
startTime = (datetime.now() - timedelta(days=30)).date().strftime("%Y-%m-%d")
date_list = []
df_list = []
allDates = pd.date_range(start=startTime, end = endTime)
for oneDate in allDates:
    oneDate = oneDate.date().strftime("%Y-%m-%d")
    oneDayData = auth2_client.intraday_time_series('activities/heart', base_date=oneDate, detail_level='1sec')
    df = pd.DataFrame(oneDayData['activities-heart-intraday']['dataset'])
    date_list.append(oneDate)
    df_list.append(df)
    
final_df_list = []
for date, df in zip(date_list, df_list):
    if len(df) == 0:
        continue
    df.loc[:, 'date'] = pd.to_datetime(date)
    final_df_list.append(df)

df_heartRate = pd.concat(final_df_list, axis = 0)
df_heartRate['datetime'] = pd.to_datetime(df_heartRate['date']) +pd.to_timedelta(df_heartRate['time'])
display(df_heartRate)
```

## 8.) Get Sleep Data

```python
startTime = pd.datetime(year = 2020, month = 1, day = 5)
endTime = pd.datetime.today().date() - datetime.timedelta(days=1)
```

```python
allDates = pd.date_range(start=startTime, end = endTime)
```

```python
date_list = []
df_list = []
stages_df_list = []

allDates = pd.date_range(start=startTime, end = endTime)

for oneDate in allDates:
    
    oneDate = oneDate.date().strftime("%Y-%m-%d")
    
    oneDayData = auth2_client.sleep(date=oneDate)
    
    # get number of minutes for each stage of sleep and such. 
    stages_df = pd.DataFrame(oneDayData['summary'])

    df = pd.DataFrame(oneDayData['sleep'][0]['minuteData'])
    
    date_list.append(oneDate)
    
    df_list.append(df)
    
    stages_df_list.append(stages_df)
    
final_df_list = []

final_stages_df_list = []

for date, df, stages_df in zip(date_list, df_list, stages_df_list):

    if len(df) == 0:
        continue
    
    df.loc[:, 'date'] = pd.to_datetime(date)
    
    stages_df.loc[:, 'date'] = pd.to_datetime(date)
    
    final_df_list.append(df)
    final_stages_df_list.append(stages_df)

final_df = pd.concat(final_df_list, axis = 0)

final_stages_df = pd.concat(final_stages_df_list, axis = 0)
```

```python
columns = final_stages_df.columns[~final_stages_df.columns.isin(['date'])].values
```
